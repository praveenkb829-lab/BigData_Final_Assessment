```python
# Load the cleaned table
df = spark.table("video_game_sales_clean")

display(df.limit(10))
df.printSchema()
print("Rows:", df.count())
```

```python
numeric_cols = ["Year", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]

display(df.select(numeric_cols).describe())
```

```python
from pyspark.sql import functions as F

sales_by_platform = (
    df.groupBy("Platform")
      .agg(F.round(F.sum("Global_Sales"), 2).alias("Total_Global_Sales"))
      .orderBy(F.desc("Total_Global_Sales"))
)

display(sales_by_platform)
```

```python
sales_by_genre = (
    df.groupBy("Genre")
      .agg(F.round(F.sum("Global_Sales"), 2).alias("Total_Global_Sales"))
      .orderBy(F.desc("Total_Global_Sales"))
)

display(sales_by_genre)
```
